```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
pre_treatment = pd.read_csv('data/Romain_pre.csv')
post_treatment = pd.read_csv('data/Romain_post.csv')
pre_treatment.drop(post_treatment[post_treatment['Day Type'] == '0: All Days (M-Su)'].index,inplace=True)
pre_treatment.drop(post_treatment[post_treatment['Day Part'] == '0: All Day (12am-12am)'].index,inplace=True
post_treatment.drop(post_treatment[post_treatment['Day Type'] == '0: All Days (M-Su)'].index,inplace=True)
post_treatment.drop(post_treatment[post_treatment['Day Part'] == '0: All Day (12am-12am)'].index,inplace=True)

post_speeds = post_treatment['Avg Spot Speed (mph)']

t_stat, p_value = stats.ttest_rel(pre_speeds, post_speeds)

df = pd.DataFrame({   
    'pre_speeds': pre_speeds,
    'post_speeds': post_speeds
})

summary_table = df.describe()
print(summary_table) 
bar_colors = ['red','blue']
plt.figure(figsize=(6, 4))
plt.bar(['Pre-Treatment', 'Post-Treatment'], 
        [pre_speeds.mean(), post_speeds.mean()], 
        yerr=[pre_speeds.std(), post_speeds.std()], capsize=5,color = bar_colors)

plt.ylabel('Average Spot Speed (mph)')
plt.title('Oakwood Traffic Circle: Pre vs Post-Treatment Speeds')
plt.show()
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")
```
